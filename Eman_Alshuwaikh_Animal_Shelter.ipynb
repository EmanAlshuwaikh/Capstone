As shown before, the same colors are there but flipped to easily categorize with no duplicates. Hence, a function was created in order to do that.

```python
#Create a color function that would manage to make one category for each color or mix of color.
def color(word):
  lists = word.split('/')
  lists = list(set(lists))
  lists.sort()
  return ' - '.join(lists)
```

```python
intakes['Color'] = intakes['Color'].apply(color)
outcomes['Color'] = outcomes['Color'].apply(color)
```

I have created a function to find out the true age. However, since the function resulted in some unexpected negative value. I have decided to drop them as they are only 5 rows and will not affect the outcome. For time efficiency purposes. Also, we could not depend on the date of birth as it showcased to be uncredible.

```python
len(join[join['Date of Birth'].dt.date > join['Date_In']])
```

```python
#Trying to make the negative age to their true age using the age on intake and dates to find the outcomes age.
def old_age(age):
    if age['Age upon Outcome'] == '-1 years' or age['Age upon Outcome'] == '-2 years' or age['Age upon Outcome'] == '-3 years':
        new_age = int(age['Age upon Intake'][0]) + int((age['Date_Out'] - age['Date_In']).days / 365)
        return f'{new_age} years'
    else:
        return age['Age upon Outcome']


join['Age upon Outcome'] = join.apply(old_age, axis=1)
```

```python
join.drop(join[join['Age upon Outcome'].isin(['-1 years', '-2 years'])].index, inplace=True)
```

I wanted to compare the animals types outcomes for the top 3 to each other.

```python
#Plot three pie graphs in the same output
animal_types = ['Dog', 'Cat', 'Other']
legends = []

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
#Loop is used to go over the data for each animal type once, extract the count of animal in each outcome, create the plot and go over the other types after.
for i, animal_type in enumerate(animal_types):

    animal = outcomes[outcomes['Animal Type'] == animal_type].copy()


    top_outcomes = animal['Outcome Type'].value_counts().index[:4]


    animal.loc[~animal['Outcome Type'].isin(top_outcomes), 'Outcome Type'] = 'Other'


    ax = axes[i]
    ax.pie(animal['Outcome Type'].value_counts().values,
           labels=animal['Outcome Type'].value_counts().index,
           autopct='%1.1f%%',
           startangle=190,
           textprops={'color': 'black'})
    ax.axis('equal')
    ax.set_title(f'{animal_type}s', fontsize=14)

    legends.append(animal['Outcome Type'].value_counts().index)


legend_labels = [i for x in legends for i in x]


plt.legend(legend_labels, loc="best")
fig.suptitle('Outcome by Animal Type', fontsize=20)
plt.tight_layout()
#plt.savefig('Outcome_by_Animal.png')
plt.show()

```

###Average Length of Stay

I wanted to assess the length of stay but noticed that there were negative values and that will affect the analysis.


```python
# The function 'lostay' excludes negative values from the dates and find the average.The issue occurred after joining the tables.

#Create a new dataframe for the function.
stay = pd.DataFrame({'Animal ID': join['Animal ID'], 'Date_In': join['Date_In'], 'Date_Out': join['Date_Out'], 'Outcome_Type': join['Outcome Type']})

#change their types.
stay['Date_In'] = pd.to_datetime(stay['Date_In'], format='%m/%d/%Y')
stay['Date_Out'] = pd.to_datetime(stay['Date_Out'], format='%m/%d/%Y')

#Creating the function and adding an outcome type to filter.
def lostay(stay, outcome_type=None):
  #To only choose the outcome type that we specify
    if outcome_type:
        type_stay = stay[stay['Outcome_Type'] == outcome_type].copy()
    else:
        type_stay = stay.copy()

    total_shelter_days = 0
    total_animals = type_stay['Animal ID'].count()
#Change the type to days.
    type_stay['Length_Of_Stay'] = (type_stay['Date_Out'] - type_stay['Date_In']).dt.days
#The length of stay is a positive value
    type_stay = type_stay[type_stay['Length_Of_Stay'] >= 0]
#Add the length of stay
    total_shelter_days = type_stay['Length_Of_Stay'].sum()
#calculating the average
    avg_lostay = total_shelter_days / total_animals if total_animals > 0 else 0
    return avg_lostay
```

```python
#Calculating and assigning each outcome type and their length of stay.
alos = lostay(stay)
alos_adopted = lostay(stay, outcome_type='Adoption')
alos_trans = lostay(stay, outcome_type='Transfer')
alos_rto = lostay(stay, outcome_type='Return to Owner')
alos_kill = lostay(stay, outcome_type='Euthanasia')

print(f'Average Length of Stay: {alos:.2f} days')
print(f'Average Length of Stay for Adoption: {alos_adopted:.2f} days')
print(f'Average Length of Stay for Transfer: {alos_trans:.2f} days')
print(f'Average Length of Stay for Return to Owner: {alos_rto:.2f} days')
print(f'Average Length of Stay for Euthanasia: {alos_kill:.2f} days')
```

```python
#Plotting the bar graph to showcase the average days of stay for each outcome type using the results of the function above.
outcomes = ['All Outcomes', 'Adoption', 'Transfer', 'Return to Owner', 'Euthanasia']
alos_values = [alos, alos_adopted, alos_trans, alos_rto, alos_kill]

plt.figure(figsize=(10, 6))
plt.bar(outcomes, alos_values, color=['pink', 'lightgreen', 'lightblue', 'purple', 'darkgrey'])
plt.title('Average Length of Stay (Days)')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=45)
plt.tight_layout()
#plt.savefig('alos.png')
plt.show()
```
